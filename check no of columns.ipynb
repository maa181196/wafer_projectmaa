```python
def checking_no_of_column(source):
    lg.basicConfig(filename='no_of_columns.log',level=lg.INFO, format ='%(asctime)s %(message)s')
    lg.info("the exceution will be start from here",source)
    count=0
    try:
        for i in os.listdir("good_data1"):
            lg.info("we are going to exceute")
            a=source+"\\"+'good_data1\\'+i
            df =pd.read_csv(a)
            if df.shape[1]!=592:
                b=source+"\\"+'bad_data1'
                shutil.move(a,b)
                count+=1
            else:
                pass
        lg.info("total no of file transfer into bad_data",count)
    except Exception as e:
        lg.info("there is error",e)
        
        
    
```
